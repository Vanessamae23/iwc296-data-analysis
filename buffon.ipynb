```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

$$ x \sin U(0, D/2) $$
$$ \theta \sin U(0, \pi / 2) $$

```python
D, L = 4, 2
M, n = 1000, 2000
prob = [0] * M
```

```python
for j in range(M):
    cnt = 0 # Getting the estimate n times
    for i in range(n):
        # Generate the random variables
        x = np.random.uniform(0, D/2)
        theta = np.random.uniform(0, np.pi / 2)

        # Condition for the needle to lie across a line
        if x < (np.sin(theta) * L/2):
            cnt += 1
    prob[j] = cnt / n

plt.hist(prob, color='red', edgecolor='white', bins=20)
print(f"The probability value is {np.mean(prob):.4f} and std {np.std(prob)}")
estPi = (2 * L) / (np.mean(prob) * D)
print(f"The estimated PI (should be close to the real value of PI): {np.mean(estPi):.4f}")
```

```python
xvector = [0] * n
yvector = [0] * n
pi = [0] * M

for j in range(M):
    cnt = 0
    x = np.random.uniform(-1, 1, n)
    y = np.random.uniform(-1, 1, n)
    for i in range(n):
        if (x[i] ** 2) + (y[i] ** 2) < 1:
            cnt += 1
    pi[j] = 4 * (cnt / n)

print(f"The estimated pi is {np.mean(pi):.4f} and std is {np.std(pi):.4f}")
plt.hist(pi, color='red', edgecolor='white', bins=20)
```
