```python
import numpy as np
import math
import matplotlib.pyplot as plt
import sympy as sp
from sympy import *
```

#### Summation

```python
def sum(x):
    sum = 0
    for value in x:
        sum  += value
    return sum
    
v = [2,3,4,10,24]
sum(v)
```

#### Mean

```python
def mean(x):
    sum = 0
    for value in x:
        sum  += value
    return sum/len(x)
    
v = [2,3,4,6]
mean(v)
```

#### Max

```python
def maxf(x):
    max = x[0]
    for value in x:
        if value >= max:
            max = value
    return max

maxf([2,3,4,5,-7])
```

#### Standard Deviation

```python
def std(x):
    std=np.sqrt(np.sum((x-np.mean(x))**2)/(len(x)-1))
    return std
v = [2,3,4,7]
round(std(v),4)
```

#### Correlation

```python
def correlation(x,y):
    if len(x) != len(y):
        print("The number of observations of two vectors should be the same")
    else:   
        Sxx=np.sum((x-np.mean(x))**2) 
        Syy=np.sum((y-np.mean(y))**2) 
        Sxy=np.sum((x-np.mean(x))*(y-np.mean(y))) 
        return Sxy/np.sqrt(Sxx*Syy)

a = [1,3,4,7,8]
b = [3,4,5,9,7]

round(correlation(a,b),3)
```

#### Angle between two lines

```python
def angle(m1,m2):
    cos_theta = (1+m1*m2)/(sqrt(1+m1**2)*sqrt(1+m2**2))
    theta = math.acos(cos_theta)
    angle_degree = theta*180/(math.pi)
    return min(angle_degree, 180-angle_degree)

angle(2,-1/2)
```

#### Quadratic Formula

```python
def quadeq(a,b,c):
    D=b**2-4*a*c
    if D<0:
        Re1 = -b/(2*a)
        Im1 =  np.sqrt(-D)/(2*a)
        Im2 = -np.sqrt(-D)/(2*a)
        print("we have two complex roots:")
        print(complex(Re1,Im1),",",complex(Re1,Im2))
    elif D==0:
        x = -b/(2*a)
        print("we have one real root:")
        print(x)            
    else:
        x1 = -b/(2*a)+np.sqrt(D)/(2*a)
        x2 = -b/(2*a)-np.sqrt(D)/(2*a)
        print("we have two real roots:")
        print(x1,",", x2)

quadeq(1,-2, 5)
quadeq(1,-1,-2)
quadeq(1,-2, 1)
```

#### Quartiles

```python
x = [2,3,4,7,10,15,16,18,25,28,30,34,50]

def quartiles(x):
    n = len(x)
    m = (n+1)/2
    h1 = (m+1)/2
    h3 = n+1-h1
    m_int = math.floor(m)
    m_dec = m-m_int
    median = x[m_int-1]*(1-m_dec)+x[m_int]*m_dec
    
    h1_int = math.floor(h1)
    h1_dec = h1-h1_int
    q1 = x[h1_int-1]*(1-h1_dec)+x[h1_int]*h1_dec

    h3_int = math.floor(h3)
    h3_dec = h3-h3_int
    q3 = x[h3_int-1]*(1-h3_dec)+x[h3_int]*h3_dec
    
    return [q1, median, q3]

quartiles(x)
```

#### Factorial

```python
def Factorial(n):
    result=1
    for i in range(1,n+1):
        result=result*i
    return result

Factorial(6)
```

#### Range

```python
def checkrange(a,b,n):
    if n<a | n>b:
        print(n, "is in NOT the interval [",a,",",b,"]")
    else:
        print(n, "is in the interval [",a,",",b,"]")

checkrange(3,8,5)
checkrange(3,8,10)
```

#### Newton - Raphson

```python
def derivative(func, x, dx=1e-6):
    return (func(x + dx) - func(x - dx)) / (2 * dx)

def newtonRaphson(f, x0, tol=1e-6, iter=1000):
    x = x0
    for i in range(iter):
        if abs(f(x)) < tol:
            return x
        x = x - f(x) / derivative(f, x, dx=1e-6)
ax = plt.axes()
x = np.linspace(-5, 3, 100)
ax.plot(x, np.exp(x) - 2*x - 3, color='r')
plt.axhline(y=0, color='b', linestyle='--')
```

```python
def f(x):
    return math.exp(x) - 2*x - 3

root = newtonRaphson(f, 3)
print(round(root,3))
ax = plt.axes()
x = np.linspace(-5, 3, 100)
ax.plot(x, np.exp(x) - 2*x - 3, color='r')
plt.axhline(y=0, color='b', linestyle='--')
plt.axvline(x=root, color='g', linestyle='--')
```
