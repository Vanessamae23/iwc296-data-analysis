```python
from datetime import datetime
import pandas as pd

year = input("Insert your year here")
month = input("Enter your month of birthdate")
date = input("Enter your date of birth")

DAYS_IN_YEAR = 365.2425
birthdate = datetime(int(year), int(month), int(date))

days = int((datetime.today() - birthdate).days)
age = days / DAYS_IN_YEAR
print(f"Your age is {age} ({days} days)")
```
