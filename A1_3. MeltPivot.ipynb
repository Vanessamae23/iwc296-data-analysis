```python
lang_long = pd.read_csv("/Users/vanessamae/Documents/playground/python/region_lang_top5_cities_long.csv")
lang_long
```

```python
lang_home_tidy = lang_long.pivot(
    index=["region", "category", "language"],
    columns=["type"],
    values=["count"]
).reset_index()
display(lang_home_tidy)
display(lang_home_tidy[["count"]])
display(lang_home_tidy[[("count", "most_at_home")]])
```

```python
lang_home_tidy.columns = [
    "region",
    "category",
    "language",
    "most_at_home",
    "most_at_work",
]
display(lang_home_tidy)
display(lang_home_tidy[["most_at_home"]])
```

```python
lang_messy = pd.read_csv("data/region_lang_top5_cities_messy.csv")
lang_messy
```

```python
lang_messy_longer = lang_messy.melt(
    id_vars=["category", "language"],
    var_name="region",
    value_name="value",
)

lang_messy_longer
```

```python
tidy_lang = lang_messy_longer.drop(columns=["value"])
tidy_lang[["most_at_home", "most_at_work"]] = lang_messy_longer["value"].str.split("/", expand=True)
tidy_lang
```

```python
tidy_lang.info()
```

```python
tidy_lang["most_at_home"] = tidy_lang["most_at_home"].astype("int")
tidy_lang["most_at_work"] = tidy_lang["most_at_work"].astype("int")
tidy_lang
```
