### **Uniform Distribution**
The **Uniform Distribution** is a continuous probability distribution where all outcomes in an interval are equally likely.

##### Probability Density Function (PDF)

For a continuous uniform distribution on the interval \([a, b]\):
$$
f(x) =
\begin{cases}
\dfrac{1}{b - a}, & a \leq x \leq b \\
0, & \text{otherwise}
\end{cases}
$$

#### Key Properties

- **Support:** $ x \in [a, b] $
- **Height of PDF:** $ \dfrac{1}{b - a} $
- **Shape:** Constant (flat) over $[a, b]$

#### Expected Value (Mean) and Variance
$
{E}[X] = \dfrac{a + b}{2}
$ $\quad$ and $\quad$
$
\text{Var}(X) = \dfrac{(b - a)^2}{12}
$

#### Other Features

- **Median:** $ \dfrac{a + b}{2} $ (same as mean)
- **Mode:** Any value in $[a, b]$ (all equally likely)
- **Standard Deviation:** $ \sigma = \dfrac{b - a}{\sqrt{12}} $
- **CDF (Cumulative Distribution Function):**
$
F(x) =
\begin{cases}
0, & x < a \\
\dfrac{x - a}{b - a}, & a \leq x \leq b \\
1, & x > b
\end{cases}
$

```python
# Parameters
a, b = 0, 1
x = np.linspace(a-.3, b+.3, 200)  # restrict to [a, b]

# pdf - f(x) 
pdf = np.where((x >= a) & (x <= b), 1 / (b - a), 0)

# graph of f(x) 
plt.figure(figsize=(6, 4))
plt.plot(x, pdf, label='Uniform PDF', color='blue')

plt.title('Uniform Distribution PDF')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-0.02, 1.6)
plt.show()
```

```python
# Random number generation
x = np.random.uniform(a, b, n)
prob = len(x[x > 0.3]) / n
print(prob)

plt.hist(x, color="blue", edgecolor='white', bins=10)
plt.show()
```


### **Normal Distribution**

The **Normal Distribution** (also called Gaussian Distribution) is one of the most important continuous probability distributions in statistics. It is widely used in natural and social sciences due to the Central Limit Theorem.



##### Probability Density Function (PDF)

For a normal distribution with mean $\mu$ and variance $\sigma^2$:

$
f(x) = \dfrac{1}{\sqrt{2 \pi \sigma^2}} \exp\left( -\dfrac{(x - \mu)^2}{2\sigma^2} \right), \quad -\infty < x < \infty
$



#### Key Properties

- **Support:** $ x \in (-\infty, \infty) $
- **Shape:** Bell-shaped, symmetric about $\mu$
- **Parameters:**  
  - Mean ($\mu$): center of the distribution  
  - Variance ($\sigma^2$): spread of the distribution  



#### Expected Value (Mean)

$
\mathbb{E}[X] = \mu
$ $\quad$ and $\quad$
$
\text{Var}(X) = \sigma^2
$



#### **Other Features**

- **Median:** $\mu$ (same as mean)
- **Mode:** $\mu$
- **Standard Deviation:** $\sigma$
- **Skewness:** 0 (perfectly symmetric)
- **Kurtosis:** 3 (mesokurtic)
- **CDF (Cumulative Distribution Function):**
$
F(x) = \dfrac{1}{2} \left[ 1 + \text{erf}\left( \dfrac{x - \mu}{\sigma \sqrt{2}} \right) \right]
$

```python
# Parameters
mu, sigma = 0, 1
mu1, sigma1 = 4, 2
x = np.linspace(-5, 11, 200)

# pdf - f(x) 
pdf = norm.pdf(x, mu, sigma)
pdf1 = norm.pdf(x, mu1, sigma1)

# graph of f(x) 
plt.figure(figsize=(6,4))
plt.plot(x, pdf, color='blue')
plt.plot(x, pdf1, color='red')
plt.title('Normal distribution pdf')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# Random number generation
mu = 4
sigma = 2
n = 10000

x = np.random.normal(mu, sigma, n)
plt.hist(x,color='orange', bins = 20, edgecolor='white')
plt.show()
```

```python
# Pr(1<X<6)
prob = len(x[(x > 1) & (x < 6)])/n
print(prob)
```


### **Gamma Distribution**

The **Gamma Distribution** is a continuous probability distribution often used to model waiting times and is a generalization of the exponential distribution.


#### Probability Density Function (PDF)

For a Gamma distribution with shape parameter $\alpha > 0$ and rate parameter $\beta > 0$:

$
f(x; \alpha, \beta) =
\begin{cases}
\dfrac{1}{\Gamma(\alpha)\beta^\alpha} x^{\alpha - 1} e^{-\frac{x}{\beta}}, & x > 0 \\
0, & \text{otherwise}
\end{cases}
$

where $\Gamma(\alpha)$ is the Gamma function:
$
\Gamma(\alpha) = \displaystyle\int_0^\infty t^{\alpha - 1} e^{-t} \, dt
$


#### Key Properties

- **Support:** $ x > 0 $
- **Shape:** Right-skewed (depends on $\alpha$)
- **Parameters:**  
  - Shape ($\alpha$): controls skewness  
  - Rate ($\beta$): controls scale  

#### Expected Value (Mean) and Variance

$
\mathbb{E}[X] = {\alpha}{\beta}
$ $\quad$
and $\quad$
$
\text{Var}(X) = {\alpha}{\beta^2}
$



#### Other Features

- Mode
$
\begin{cases}
(\alpha - 1){\beta}, & \alpha \ge 1 \\
0, & \alpha < 1
\end{cases}
$

- Standard Deviation
$
\sigma = \sqrt{{\alpha}{\beta^2}}
$

- Special Cases
  - If $\alpha = 1$, Gamma becomes **Exponential** distribution.
  - If $\alpha$ is an integer, Gamma is related to Erlang distribution.

```python
# Parameters
alpha, beta = 2, 1
alpha1, beta1 = 3, 2
x = np.linspace(-5, 11, 200)

# pdf - f(x) 
pdf = gamma.pdf(x, alpha, beta)
pdf1 = gamma.pdf(x, alpha1, beta1)

# graph of f(x) 
plt.figure(figsize=(6,4))
plt.plot(x, pdf, color='blue')
plt.plot(x, pdf1, color='red')
plt.grid
plt.title('Gamma distribution pdf')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# Random number generation
alpha = 2
beta = 1
n = 10000

x = np.random.gamma(alpha, beta, n)
plt.hist(x,color='orange', bins = 20, edgecolor='white')
plt.show()
```

```python
# Pr(X<2)
len(x[x < 2]) / n
```

```python
# Pr(X^2-Y>10) with X ~ N(4,3^2) Y ~ N(10, 7^2) 
mu1=1
mu2=10
sigma1=3
sigma2=7

M = 1000
probabilities = [0] * M
for j in range(M):
    x = np.random.normal(mu1, sigma1, n)
    y = np.random.normal(mu2, sigma2, n)
    z = x**2 - y
    prob = len(z[z>10])/n
    probabilities[j] = prob

print(probabilities)
plt.hist(probabilities, color='brown', edgecolor='white')
print(f"Mean: {round(np.mean(probabilities), 3)}")
print(f"STD: {round(np.std(probabilities), 3)}")
```
