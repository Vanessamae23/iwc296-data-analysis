```python
import matplotlib.pyplot as plt
import numpy as np

n = 10000
ax = plt.axes()
x = np.linspace(0,1,1000)
ax.plot(x, 3*x**2, color='red')
plt.show()
```

```python
u = np.random.uniform(0,1,n)
x1 = u**(1/3)
plt.hist(x1, color='red', edgecolor='white', bins=20)
```

```python
# graph of f(x)
n = 10000
ax = plt.axes()
x = np.linspace(0,1,1000)
ax.plot(x, np.sin(x), color='red')
plt.show()
```

```python
u = np.random.uniform(0,1,n)
x2 = np.arccos(1-u) # inverse of F()
plt.hist(x2, color='red', edgecolor='white', bins=20) # histogram of the generated rv
```

```python
# graph of f(x)
n = 10000
ax = plt.axes()
x = np.linspace(1,2,1000)
ax.plot(x, (1/6) * x + 0.5, color='red')
plt.show()
```

```python
# ((x ** 2 + 3x)/6) - 2/6
# 6u = x**2 + 3x -2
# x**2 + 3x -2 - 6u = 0
# (-3 + sqrt(9 - 4(1)(-2 - 6u))) / 2
u = np.random.uniform(0,1,n)
x3 = (-3 + np.sqrt(9 - 4 * (-2 - 6 * u))) / 2
plt.hist(x3, color='red', edgecolor='white', bins=20) # histogram of the generated rv
```

```python
# graph of f(x) = np.exp(-3x) / 3
# np.exp(-3x) / (-9)
n = 10000
ax = plt.axes()
x = np.linspace(1,2,1000)
ax.plot(x, np.exp(-1 / 3 * x) / 3, color='red')
plt.show()
```

```python
# graph of f(x) = np.exp(-1/3x) / 3

u = np.random.uniform(0,1,n)
ln = np.log # Create an alias
x4 = (-3) * ln(1- u) 
plt.hist(x4, color='red', edgecolor='white', bins=20) # histogram of the generated rv
```
