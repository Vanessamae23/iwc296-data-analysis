#### Region Data: Region

```python
region_data = pd.read_csv("region_data.csv")
region_data
```

#### Filter Observations: isin()

```python
city_names = ["Toronto", "Montréal", "Vancouver", "Calgary", "Edmonton"]
five_cities = region_data[region_data["region"].isin(city_names)]
five_cities
```

#### Comparison of values: == vs isin()

```python
pd.Series(["Vancouver", "Toronto"]) == pd.Series(["Toronto", "Vancouver"])
```

```python
pd.Series(["Vancouver", "Toronto"]).isin(pd.Series(["Toronto", "Vancouver"]))
```

```python
region_data[region_data["population"] > 2000000]
```

```python
region_data.query("population > 2000000")
```

#### Reading Data: Tidy_lang

```python
tidy_lang = pd.read_csv("tidy_lang_data.csv")
tidy_lang
```

#### [Extracting columns by name] - filter rows and select columns at the same time using loc[]

```python
tidy_lang.loc[
    tidy_lang["region"] == "Toronto",     # filter rows
    ["language", "region", "most_at_home", "most_at_work"]  # select columns
]
```

```python
tidy_lang.loc[
    tidy_lang["region"] == "Toronto",
    "language":"most_at_work"
]
```

```python
tidy_lang.loc[10:, "region":]
```

```python
tidy_lang.loc[:15, :"region"]
```

```python
tidy_lang.loc[10:16, tidy_lang.columns.str.startswith("most")]
```

```python
tidy_lang.loc[20:30, tidy_lang.columns.str.contains("_")]
```

#### Using iloc[] to extract rows and columns by position - Python starts counting at 0, so the second column "language" has index 1.

```python
tidy_lang.iloc[:, 0]
```

```python
tidy_lang.iloc[:, 0:]
```

```python
tidy_lang.iloc[5:10, :4]
```
