```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 15,1000)
y = np.linspace(0, 15,1000)
f = np.exp(x - 5)/(1 + np.exp(x - 5))**2
g = 0.32 * np.exp(-0.32 * y)

plt.plot(x,f,color="red")
plt.plot(y,g,color="blue")
plt.legend()
plt.grid()
plt.show()
```

```python
N = 1000
u1 = np.random.uniform(0,1,N)
u2 = np.random.uniform(0,1,N)

x = 5 + np.log(u1 / (1 - u1))
y = np.log(1- u2) / (-0.32)

plt.xlim(-5, 15)
plt.hist(x, color = "red", edgecolor="white", bins=30)
plt.show()

plt.xlim(-5, 15)
plt.hist(y, color = 'blue', edgecolor = "white", bins=30)
plt.show()
```

```python
M = 1000

probX = [0] * M
probY = [0] * M
probZ = [0] * M

for j in range(M):
    u1 = np.random.uniform(0, 1, N)
    u2 = np.random.uniform(0, 1, N)
    x = 5 + np.log(u1 / (1 - u1))
    y = np.log(1- u2) / (-0.32)
    z = x - 2*y

    probX[j] = len(x[(x > 4)]) / N
    probY[j] = len(y[(y > -2) & (y < 3)]) / N
    probZ[j] = len(z[(z > 0)]) / N

plt.hist(probX, color='red', edgecolor='white', bins=30)
plt.axvline(np.mean(probX), color='black', linestyle="dashed", linewidth=2)
print(f"The estimate of the prob(x > 4) is {np.mean(probX)}")
plt.show()

plt.hist(probY, color='blue', edgecolor='white', bins=30)
plt.axvline(np.mean(probY), color='black', linestyle="dashed", linewidth=2)
print(f"The estimate of the probY is {np.mean(probY)}")
plt.show()

plt.hist(probZ, color='green', edgecolor='white', bins=30)
plt.axvline(np.mean(probZ), color='black', linestyle="dashed", linewidth=2)
print(f"The estimate of the probZ is {np.mean(probZ)}")
plt.show()
```

### Probability of Broken Stick problem

```python
M = 10000
n = 100
prob = [0] * M
prob1 = [0] * M

for j in range(M):
    cnt = 0
    cnt1 = 0
    for i in range(n):
        x = np.random.uniform(0, 1)
        y = np.random.uniform(0, 1)

        a = min(x, y)
        b = max(x, y) - a
        c = 1 - max(x, y)

        s3 = max(a, b, c)
        s1 = min(a, b, c)
        s2 = 1 - s1 - s3

        if s1 + s2 > s3:
            if (s1 ** 2) + (s2 ** 2) > (s3 ** 2):
                cnt1 += 1
            cnt += 1
    prob[j] = cnt / n
    prob1[j] = cnt1 / n

plt.xlim(0,1)
plt.hist(prob, color='orange', bins=20, edgecolor='white')

print(f"Mean for any triangle is {np.mean(prob)}")
print(f"Mean for obtuse and acute is {np.mean(prob1)}")
```
